```python
help_inequity_df_new_survey = help_inequity_df_new[help_inequity_df_new['Problem Name'].isin(['How confident are you in your knowledge of this topic? -- (Not at all confident / Extremely confident)',\
       'How would you describe your tutoring experience and skills? -- (Beginner Tutor / Expert Tutor)',\
       "Now that you've completed the module, how confident are you in your knowledge of this topic? -- (Not at all confident / Extremely confident)",\
       'The lesson was easy to understand: * -- (Disagree / Agree)',\
       'I am confident I can apply what I learned: * -- (Disagree / Agree)'])]

a_categor_map = {'How confident are you in your knowledge of this topic? -- (Not at all confident / Extremely confident)':'confident_know_pre',
'How would you describe your tutoring experience and skills? -- (Beginner Tutor / Expert Tutor)':'tutoring_experience_pre',
"Now that you've completed the module, how confident are you in your knowledge of this topic? -- (Not at all confident / Extremely confident)":'confident_know_post',
'The lesson was easy to understand: * -- (Disagree / Agree)':'lesson_understanding_post',
'I am confident I can apply what I learned: * -- (Disagree / Agree)':'confident_apply_post'}
help_inequity_df_new_survey.loc[:, 'problem_type'] = help_inequity_df_new_survey['Problem Name'].replace(a_categor_map)
```

```python

help_inequity_fliter_v1_new = help_inequity_df_new[help_inequity_df_new['Problem Name'].isin(list(add_help_inequity_dict.keys()))]
help_inequity_fliter_v1_new = help_inequity_fliter_v1_new[~help_inequity_fliter_v1_new['Input'].isna()]
help_inequity_fliter_v2_new = help_inequity_fliter_v1_new[help_inequity_fliter_v1_new['Input'].astype(str).str.len() > 3]
help_inequity_fliter_v2_new.loc[:, 'All_outcome'] = help_inequity_fliter_v2_new['Outcome'].fillna(help_inequity_fliter_v2_new['Source of Human Truth [what Sanjit and Shambhavi agree is the source truth]'])
help_inequity_fliter_v2_new.loc[:, 'Problem type'] = help_inequity_fliter_v2_new['Problem Name'].replace(add_help_inequity_dict)
```

```python
# Filter the DataFrame to include only specific 'Problem type' values
mask = help_inequity_fliter_v2_new['Problem type'].isin(['explained_mcq_pre', 'explained_mcq_post', 'explained_open_post', 'explained_open_pre'])

# Update the 'Problem type' column in the original DataFrame where the condition is met
help_inequity_fliter_v2_new.loc[mask, 'Problem type'] = (
    help_inequity_fliter_v2_new.loc[mask, 'Level (Level2)'].str.split('_').str[-1] + 
    '_' + 
    help_inequity_fliter_v2_new.loc[mask, 'Problem type']
)
```
